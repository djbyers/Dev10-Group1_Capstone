```python
#first, I load in my dataset from the sql database and process it for ML use
import pymssql
import pandas as pd
import numpy as np
database = "group1"
table = "dbo.capstone_sql_table"
username = "group1user"
password  = "everythingIsAwesome!"
server = "database2108.database.windows.net"

conn = pymssql.connect(server,username, password,database)

cursor = conn.cursor()
query = 'select * from capstone_sql_table'
rawDF = pd.read_sql(query, conn)
```

```python
no_giniDF = rawDF[rawDF['Gini index (World Bank estimate)'].isna()]
no_giniDF.head()
```

```python
no_giniDF.drop(columns=['timestamp','Poverty headcount ratio at national poverty lines (% of population)', 
                     'Life expectancy at birth, total (years)', 
                     'Mortality rate, infant (per 1,000 live births)', 'Unemployment, total (% of total labor force) (modeled ILO estimate)', 
                     'Income share held by highest 10%', 'Central government debt, total (% of GDP)', 
                     'International migrant stock, total', 'Stocks traded, turnover ratio of domestic shares (%)'], inplace=True)
```

```python
no_giniDF.info()
```

```python
no_giniDF = no_giniDF[no_giniDF['Stocks traded, total value (% of GDP)'].notna()]
no_giniDF = no_giniDF[no_giniDF['Inflation, consumer prices (annual %)'].notna()]
no_giniDF = no_giniDF[no_giniDF['Tax revenue (% of GDP)'].notna()]
no_giniDF = no_giniDF[no_giniDF['GDP per capita (current US$)'].notna()]
no_giniDF = no_giniDF[no_giniDF['Broad money (% of GDP)'].notna()]
no_giniDF = no_giniDF[no_giniDF['Total reserves (includes gold, current US$)'].notna()]
no_giniDF = no_giniDF[no_giniDF['Exports of goods and services (% of GDP)'].notna()]
no_giniDF.info()
```

```python
no_giniDF.head()
```

```python
c_y_DF = no_giniDF.drop(columns=['Gini index (World Bank estimate)'])
modelDF = c_y_DF.drop(columns=['country','year'])
# used the code up to this point to get these two dataframes into Power BI
# I then used the modelDF and the pickle file to create a predicted outcome column in Power Query
# I then merged it with the c_y_DF in Power BI to make visualizations with both the predicted outcomes and the country/year
# The rest of this file is how I would have created the outcomes column in Python rather than Power Query
```

```python
import pickle
filename = 'gini_index_final_model_with_imputation.pkl'
```

```python
loaded_model = pickle.load(open(filename, 'rb'))
predicted_gini = loaded_model.predict(modelDF)
print(predicted_gini)
```

```python
predictions = pd.DataFrame(predicted_gini,columns=['predicted_gini'])
predictions.head()
```

```python

modelDF.reset_index(inplace=True)

modelDF.head()
```

```python
modelDF.drop(columns=['index'],inplace=True)
modelDF.head()
```

```python
modelDF['Gini ML Model Predictions']=predictions
```
